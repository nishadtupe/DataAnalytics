# Data Visualization using Seaborn

```python
# Import the necessary Libraries 

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the style 

sns.set_style('whitegrid')
```

```python
# Load the Famous titanic Dataset.

titanic = sns.load_dataset('titanic')
```

```python
# Display the few samples from the dataset

titanic.head()
```

```python
# Jointplot shows the relationship between two variables and also annotes the pearson correlation coefficient.

sns.jointplot(x='fare',y='age',data=titanic)
```

```python
# distplot of seabron allows us to see the distribution of underlying vairble 

sns.distplot(titanic['fare'],bins=30,color='blue')
```

```python
# Boxplot allows use to see the statistical ranges across categories
# Here we can see the age distribution for 1st , 2nd and 3rd class

sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
# Swarm plot is different kind of plot boc plot wider the swarm more dense the variable in that range.
# So we can in 3rd class we have more people from range 20-40

sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

```python
# countplot simply shows count by category
sns.countplot(x='sex',data=titanic)
```

```python
# heatmap is excellent for seeing the correlations in the dataset.

sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```
